## Pydantic basics: Creating and Using Models

Pydantic models are the foundation of data validation in python. They use python type annotations to define the structure and validate the data at runtime. 

```python
# A base class for creating pydantic models -> BaseModel
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name: str
    age: int
    city: str
    
person1=Person(name="Sandeep", age=23, city='Hyderabad')
print(person1)
print(type(person1))
```

```python
from dataclasses import dataclass

@dataclass
class Person():
    name: str
    age: int
    city: str
    
person2=Person(name="Sandeep", age=23, city='Hyderabad')
print(person2)
print(type(person2))
```

```python
class Person(BaseModel):
    name: str
    age: int
    city: str
    
person1=Person(name="Sandeep", age=23, city=1) # ValidationError: 1 validation error for Person, Input should be a valid string [type=string_type, input_value=1, input_type=int]
print(person1)
print(type(person1))
```

```python
from dataclasses import dataclass

@dataclass
class Person():
    name: str
    age: int
    city: str
    
person3=Person(name="Sandeep", age=23, city=23)
print(person3)
print(type(person3))
```

### Model with Optional Fields

Add optional fields using python's optional type:

```python
from typing import Optional 

class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional[float]=None # optional with default None
    is_active: Optional[bool]=True # Optional with default True
```

```python
emp=Employee(id=102, name= "sandeep", department= 'IT')
print(emp)
print(type(emp))
```

```python
emp1=Employee(id=102, name= "sandeep", department= 'IT', salary=65000.23, is_active=False)
print(emp1)
```

```python
from typing import List

class Classroom(BaseModel):
    room_number: str
    students: List[str]
    capacity: int
```

```python
classroom=Classroom(room_number='342', students=['sandeep', 'gauti', 'naveen', 'siva'], capacity=23)
print(classroom)
```

## model with nested models

```python
class Address(BaseModel):
    street: str
    city: str
    country: str
class Customer(BaseModel):
    id: str
    name: str
    address: Address
    
c=Customer(id='233', name='sandeep', address={'street': 'gedge', 'city': 'kakinada', 'country': 'india'})

print(c)
```

```python
class Address(BaseModel):
    street: str
    city: str
    country: str
class Customer(BaseModel):
    id: int
    name: str
    address: Address
 
# id given str value   
c=Customer(id='233', name='sandeep', address={'street': 'gedge', 'city': 'kakinada', 'country': 'india'})

print(c)
```

### Pydantic fields : Customization and Constraints

field function in pydantic model fields beyond the basic type hints by allowing you to specify the validation rules, default values, aliases, and more. 

```python
from pydantic import BaseModel, Field
```

```python
class Item(BaseModel):
    
    # Constraints
    name: str=Field(min_length=2, max_length=49)
    price: float= Field(gt=0, lt=1000)
    quantity: int=Field(ge=0)
```

```python
item=Item(name='sandeep', price=12, quantity=23)
print(item)
```

```python
item=Item(name='sandeep', price=1100, quantity=23)
print(item)
```
